```python

import pandas as pd

# 读取 CSV 文件
data = pd.read_csv('CAISOdata.csv')

# 显示前几行数据
data.head()

```

```python
data
```

```python

# Convert 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Plot the actual demand over time
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['demand_actual'], label='Actual Demand', linewidth=0.8)
plt.title('Actual Demand Over Time')
plt.xlabel('Date')
plt.ylabel('Demand (MW)')
plt.legend()
plt.grid(True)
plt.show()
```

```python


# Convert 'date' column to datetime format if needed
data['date'] = pd.to_datetime(data['date'])

# Group the data by 'month' and calculate the average demand per month
monthly_demand = data.groupby('month')['demand_actual'].mean()

# Plot the average actual demand per month
plt.figure(figsize=(10, 6))
plt.plot(monthly_demand.index, monthly_demand.values, marker='o', linestyle='-', linewidth=2)
plt.title('Average Actual Demand Per Month')
plt.xlabel('Month')
plt.ylabel('Average Demand (MW)')
plt.xticks(range(1, 13))  # Setting x-ticks for each month (1 to 12)
plt.grid(True)
plt.show()
```

```python


# Convert 'date' column to datetime format if needed
data['date'] = pd.to_datetime(data['date'])

# Define the renewable energy supply columns we're analyzing
renewable_supply_sources = ['solar_MW', 'wind_MW', 'geothermal_MW', 'biomass_MW', 
                            'biogas_MW', 'sm_hydro_MW', 'battery_MW']

# Check if all these columns exist in the dataset
available_columns = [col for col in renewable_supply_sources if col in data.columns]

# Group data by 'month' and calculate the average supply for each available renewable source
renewable_supply_data = data[['month'] + available_columns].groupby('month').mean()

# Plot the average monthly supply from these renewable energy sources
plt.figure(figsize=(10, 6))
for source in available_columns:
    plt.plot(renewable_supply_data.index, renewable_supply_data[source], marker='o', linestyle='-', label=source)

plt.title('Average Monthly Supply from Renewable Energy Sources')
plt.xlabel('Month')
plt.ylabel('Power Supply (MW)')
plt.legend()
plt.grid(True)
plt.show()
```

```python


# Convert 'date' column to datetime format if needed
data['date'] = pd.to_datetime(data['date'])

# Define the renewable energy supply columns we're analyzing
renewable_supply_sources = ['solar_MW', 'wind_MW', 'geothermal_MW', 'biomass_MW', 
                            'biogas_MW', 'sm_hydro_MW', 'battery_MW']

# Check if all these columns exist in the dataset
available_columns = [col for col in renewable_supply_sources if col in data.columns]

# Calculate the total supply for each type of renewable energy source across the entire dataset
total_supply_per_type = data[available_columns].sum()

# Plot the total supply for each type of renewable energy source
plt.figure(figsize=(10, 6))
total_supply_per_type.plot(kind='bar', color='skyblue')
plt.title('Total Supply of Each Renewable Energy Type')
plt.xlabel('Renewable Energy Type')
plt.ylabel('Total Supply (MW)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```
